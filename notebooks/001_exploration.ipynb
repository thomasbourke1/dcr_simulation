```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve
```

```python
def calculate_N_tr(c, M_g):
    # average number of carriers trapped after a current pulse
    N_tr = c * M_g / (1-c)
    return N_tr


def calculate_tau_star(M_0, GB):
    # Effective transit time of carriers (not sure from where)
    return M_0 / (2 * np.pi * GB)

def calculate_P_d(P_d):
    # Calculates dark count probability per pulse. I_DM is primary dark current and the main thing we might vary


    P_a = 0.1 # Avalanche probability (could also find from McIntyre model)
    I_DM = 0.1e-12 # primary dark current
    tau = 2e-9 # Gate pulse width
    tau_d = 200e-9 # Trap time constant
    ELECTRONIC_CHARGE = 1.602e-19 # Electronic charge
    M_0 = 10 #DC gain
    GB = 30e9 # Gain bandwidth 
    c= 0.01 # Radio fo trapped carriers to the total carriers generated per avalanche pulse
    M_g = 1e8 # Gain when gate applied I think?
    delta_T = 1e-5 # Reciprocal of pulse repetiion rate 1GHz = 1ns

    P_d = 1 - np.exp(
        -P_a * ( (I_DM * tau / ELECTRONIC_CHARGE) + (I_DM * M_0**2 / (2* np.pi * GB))
                + (P_d * calculate_N_tr(c, M_g) * (np.exp(tau / tau_d)-1)/(np.exp(delta_T / tau_d)-1)) 
                + (P_d * calculate_N_tr(c, M_g) * (np.exp(calculate_tau_star(M_0, GB) / tau_d)-1/np.exp(delta_T / tau_d)-1) ) )
    )

    return P_d


def g(P_d):
    return calculate_P_d(P_d) - P_d

P_d_initialGuess = 1e-4

P_d_solution = fsolve(g, P_d_initialGuess)

print("P_d =", P_d_solution[0])
```
