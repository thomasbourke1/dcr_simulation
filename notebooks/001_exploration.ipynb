```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def P_d_DarkCountProbability(N_d: float, P_a: float) -> float:
    """Calculate dark count probability following Poissonian statistics

    Args:
        N_d (float): Average number of dark carriers in the multiplication region
        P_a (float): Probability of triggereing the avalanche event above the threshold gain M_g

    Returns:
        float: _description_
    """
    P_d = 1 - np.exp(- N_d * P_a)
    return  P_d
```
